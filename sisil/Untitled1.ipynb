```python
import socket

# Fungsi untuk memproses permintaan dari klien
def handle_client(client_socket):
    # Kirim respons HTTP awal untuk streaming MJPEG
    client_socket.sendall(b"HTTP/1.1 200 OK\r\n"
                          b"Server: MJPEG Server\r\n"
                          b"Content-Type: multipart/x-mixed-replace;boundary=frame\r\n\r\n")

    try:
        while True:
            # Baca data dari klien (frame MJPEG)
            frame_data = ""
            while True:
                chunk = client_socket.recv(1024)
                frame_data += chunk
                print(frame_data)
            # Kirim frame yang diterima kepada klien lain yang mengakses alamat streaming
            for client in clients:
                try:
                    client.sendall(b"--frame\r\n"
                                   b"Content-Type: image/jpeg\r\n\r\n" + frame_data + b"\r\n")
                except Exception as e:
                    # Tangani kesalahan jika ada ketika mengirimkan ke klien
                    print("Error sending frame to client:", str(e))

    except Exception as e:
        print("Error handling client:", str(e))

    finally:
        client_socket.close()

# Inisialisasi socket server
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind(("172.16.2.72", 8080))
server.listen(5)

print("Server is listening on port 8080")

clients = []

try:
    while True:
        client_socket, addr = server.accept()
        print("Accepted connection from", addr)
        clients.append(client_socket)
        handle_client(client_socket)

except KeyboardInterrupt:
    pass

finally:
    server.close()
```
