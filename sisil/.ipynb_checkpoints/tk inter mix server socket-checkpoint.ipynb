```python
#V1


from datetime import datetime
import socket
import threading
from tkinter import *
try:
    import Tkinter as tk
except ImportError:
    import tkinter as tk


a = 0
valueS = 0
valueO = 0
root = tk.Tk()
root.title("Server")
text = tk.Text(master=root)
text.pack(expand=True, fill="both")

entry = tk.Entry(master=root)
entry.pack(expand=True, fill="x")

frame = tk.Frame(master=root)
frame.pack()

w2 = Scale(root, from_=0, to=100,tickinterval=100, orient=HORIZONTAL)
w2.set(0)
w2.pack()
    
scaller_button = tk.Button(root, text="Scaller")
scaller_button.pack()

w1 = Scale(root, from_=0, to=100,tickinterval=100, orient=HORIZONTAL)
w1.set(0)
w1.pack()
    
offside_button = tk.Button(root, text="Ofset")
offside_button.pack()


def buttons():
    for i in "Connect", "ON", "OFF", "Scaler +", "Scaler -","Ofset +", "Ofset -", "Clear":
        b = tk.Button(master=frame, text=i)
        b.pack(side="left")
        yield b


b1, b2, b3, b4, b5, b6, b7, b8 = buttons()


class Server:
    clients = []

    def __init__(self):
        self.s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    def connect(self):
        self.s.bind(("", 8234))
        self.s.listen(5)
        now = str(datetime.now())[:-7]
        text.insert("insert", "({}) : Connected.\n".format(now))
        self.condition()

    def accept(self):
        c, addr = self.s.accept()
        global ADDR
        (ADDR, PORT) = addr
        now = str(datetime.now())[:-7]
        self.clients.append(c)
        print(c.getpeername())
        text.insert("insert", "({}) : {} connected.\n".format(now, str(c.getpeername())))

    def receive(self):
        for i in self.clients:
            def f():
                now = str(i.getpeername())
                data = str(i.recv(1024))[2:-1]
                if len(data) == 0:
                    print(now)
                    pass
                else:
                    print("Received data: addr={}, data={}".format(now, data))
                    text.insert("insert", "({}) : {}\n".format(now, data))

            t1_2_1 = threading.Thread(target=f)
            t1_2_1.start()

    def condition(self):
        while True:
            t1_1 = threading.Thread(target=self.accept)
            t1_1.daemon = True
            t1_1.start()
            t1_1.join(1)
            t1_2 = threading.Thread(target=self.receive)
            t1_2.daemon = True
            t1_2.start()
            t1_2.join(1)

    def send(self):
        respond = "Server: {}".format(str(entry.get()))
        print(str(entry.get()))
        now = str(datetime.now())[:-7]
        entry.delete("0", "end")
        global hex_data
        vs = format(valueS,'x')
        vo = format(valueO,'x')
        print(vs)
        print(vo)
        if a == 1 :           
            hex_data = "02140001"
        elif a == 0 :
            hex_data = "02140100"
        elif a == 2 or a == 3 :
            if valueS < 16 :
                vs = "0"+vs
            hex_data = "0105"+vs+"01"
        elif a == 4 or a == 5 :
            if valueO < 16 :
                vo = "0"+vo
            hex_data = "0105"+vo+"02"
        try:
            for i in self.clients:
#                 self.s.sendall(bytes(respond.encode("utf-8")))
                sendHex(i)
            text.insert("insert", "({}) : {}\n".format(now, respond))
        except BrokenPipeError:
            text.insert("insert", "({}) : Client has been disconnected.\n".format(now))


s1 = Server()
# List of hexadecimal values

def sendHex(i):
    binary_data = bytes.fromhex(hex_data)
    i.sendall(binary_data)
    print(binary_data)
def connect():
    t1 = threading.Thread(target=s1.connect)
    t1.start()    
    
def ON():
    global a
    a = 0
    t2 = threading.Thread(target=s1.send)
    t2.start()
    

def OFF():
#     text.delete("1.0", "end")
    global a
    a = 1
    t2 = threading.Thread(target=s1.send)
    t2.start()

def SclrPlus():
    global a
    global valueS

    if valueS < 100:
        valueS += 1
        w2.set(valueS)

    a = 2
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
def SclrMin():
    global a
    global valueS
    
    try:
        valueS -= 1
        w2.set(valueS)
    except Exception as e:
        print("Terjadi pengecualian:", e)
        
    a = 3
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
def OfstPlus():
    global a
    global valueO
    if valueO < 100:
        valueO += 1
        w1.set(valueO)
    a = 4
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
def OfstMin():
    global a
    global valueO
    
    try:
        valueO -= 1
        w1.set(valueO)
    except Exception as e:
        print("Terjadi pengecualian:", e)
    a = 5
    t2 = threading.Thread(target=s1.send)
    t2.start()

def offside_value():
    value = w1.get()  # Mendapatkan nilai slider saat ini
    global a
    global valueO

    valueO = value

    a = 4
    t2 = threading.Thread(target=s1.send)
    t2.start()
    print("Nilai slider offside:", valueO)

def scaller_value():
    value = w2.get()  # Mendapatkan nilai slider saat ini
    global a
    global valueS

    valueS = value

    a = 2
    t2 = threading.Thread(target=s1.send)
    t2.start()
    print("Nilai slider:", valueS)
    
def clear():
    text.delete("1.0", "end")

def destroy():
    root.destroy()
    exit()


if __name__ == "__main__":
    b1.configure(command=connect)
    b2.configure(command=ON)
    b3.configure(command=OFF)
    b4.configure(command=SclrPlus)
    b5.configure(command=SclrMin)
    b6.configure(command=OfstPlus)
    b7.configure(command=OfstMin)
    b8.configure(command=clear)
    offside_button.configure(command=offside_value)
    scaller_button.configure(command=scaller_value)
    t0 = threading.Thread(target=root.mainloop)
    t0.run()
```

```python
from datetime import datetime
import socket
import threading
import random
from tkinter import *
try:
    import Tkinter as tk
except ImportError:
    import tkinter as tk
 
ip_client = ""
a = 0
dev = 2
valueS = 0
valueO = 0
root = tk.Tk()
root.title("Server")
 
bframe = tk.Frame(master=root)
bframe.pack(side = "top")
 
text = tk.Text(master=root)
text.pack(expand=True, fill="both")
 
entry = tk.Entry(master=root)
entry.pack(expand=True, fill="x")
 
frame = tk.Frame(master=root)
frame.pack()
 
w2 = Scale(root, from_=0, to=100,tickinterval=100, orient=HORIZONTAL)
w2.set(0)
w2.pack()
    
scaller_button = tk.Button(root, text="Scaller")
scaller_button.pack()
 
w1 = Scale(root, from_=0, to=100,tickinterval=100, orient=HORIZONTAL)
w1.set(0)
w1.pack()
    
offside_button = tk.Button(root, text="Ofset")
offside_button.pack()
 
def buttonDev():
    for i in "Device 1", "Device 2", "Device 3", "Device 4", "Device 5","Device 6", "Device 7", "Device 8 ", "Device 9", "Device 10", "Device 11", "Device 12", "ALL", "Check Dev":
        d = tk.Button(master=bframe, text=i)
        d.pack(side="left")
        yield d
 
d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12,d13,d14 = buttonDev()
 
def buttonColors(d):
    default_color = root.cget("bg")
    d.config(bg=default_color)
        
def buttons():
    for i in "Connect", "ON", "OFF", "Scaler +", "Scaler -","Ofset +", "Ofset -", "Clear":
        b = tk.Button(master=frame, text=i)
        b.pack(side="left")
        yield b
 
b1, b2, b3, b4, b5, b6, b7, b8 = buttons()
 
 
class Server:
    clients = []
 
    def __init__(self):
        self.s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        self.lock = threading.Lock() 
        
    def connect(self):
        try:
            self.s.bind(("172.16.1.158", 8234))
            self.s.listen(5)
            now = str(datetime.now())[:-7]
            text.insert("insert", "({}) : Connected.\n".format(now))
            self.condition()
            
        except ConnectionResetError:
            print("Koneksi terputus oleh klien")
            self.reconnect()
     
    def reconnect(self):
        max_retries = 3  # Jumlah maksimum percobaan koneksi ulang
        retry_delay = 5  # Delay antara percobaan koneksi ulang (dalam detik)
        retry_count = 0
 
        while retry_count < max_retries:
            try:
                self.s.bind(("172.16.1.158", 8234))
                self.s.listen(5)
                now = str(datetime.now())[:-7]
                text.insert("insert", "({}) : Reconnected.\n".format(now))
                self.condition()
                break  # Berhasil terhubung, keluar dari loop
            except ConnectionResetError:
                print("Koneksi terputus oleh klien")
                retry_count += 1
                if retry_count < max_retries:
                    print(f"Menunggu {retry_delay} detik sebelum mencoba kembali...")
                    time.sleep(retry_delay)
                else:
                    print("Gagal terhubung setelah mencoba sebanyak yang diizinkan.")
 
            
    def accept(self):
        try:
            global c
            c, addr = self.s.accept()
            global ADDR
            (ADDR, PORT) = addr
            now = str(datetime.now())[:-7]
            self.clients.append(c)
            print(c.getpeername())
            text.insert("insert", "({}) : {} connected.\n".format(now, str(c.getpeername())))
        except ConnectionResetError:
            print("Koneksi terputus oleh klien")
            self.reconnect()
            
    def receive(self):
        for i in self.clients:
            def f():
                try:
                    now = str(datetime.now())[:-7]
                    data = str(i.recv(1024))[2:-1]
                    if len(data) == 0:
                        print(now)
                        pass
                    else:
                        print("Received data: ({}), data={}".format(now, data))
                        text.insert("insert", "({}) : {}\n".format(now, data))
                except socket.timeout:
    # Timeout tercapai
                    print("Timeout tercapai saat menerima data dari klien")
                    return
                except ConnectionResetError:
                    # Koneksi di-reset oleh peer
                    self.reconnect()
                    print("Koneksi di-reset oleh peer")
                    return
                    
            t1_2_1 = threading.Thread(target=f)
            t1_2_1.start()
 
    def condition(self):
        while True:
            try:
                t1_1 = threading.Thread(target=self.accept)
                t1_1.daemon = True
                t1_1.start()
                t1_1.join(1)
                t1_2 = threading.Thread(target=self.receive)
                t1_2.daemon = True
                t1_2.start()
                t1_2.join(1)
            except ConnectionResetError:
                print("Koneksi terputus oleh klien")
                self.reconnect()
 
    def send(self):
        
        print(str(entry.get()))
        now = str(datetime.now())[:-7]
        entry.delete("0", "end")
        global hex_data
        svr = ""
        vs = format(valueS,'x')
        vo = format(valueO,'x')
        print(vs)
        print(vo)
        if a == 1 :           
            hex_data = "02140001"
            svr = "OFF"
        elif a == 0 :
            hex_data = "02140100"
            svr = "ON"
        elif a == 2 or a == 3 :
            if valueS < 16 :
                vs = "0"+vs
            hex_data = "0105"+vs+"01"
            svr = "Control Scaller"
        elif a == 4 or a == 5 :
            if valueO < 16 :
                vo = "0"+vo
            hex_data = "0105"+vo+"02"
            svr = "Control Ofset"
        elif a == 6:
            valueC = random.randint(16, 100) 
            vc = format(valueC,'x')
            hex_data = "030415"+vc
            svr = "Check client"
        respond = "Server: {}".format(str(svr))
        try:
            with self.lock:
                for i in self.clients:
                    if i.fileno() != -1 and i.getsockopt(socket.SOL_SOCKET, socket.SO_ERROR) == 0:
                        if dev == 1 :
                            client_address = i.getpeername()[0]
                            if client_address == ip_client :
                                print("Client address :" + client_address)
                                sendHex(i)
                        elif dev == 0 :
                            sendHex(i)
                    else:
                        print("Socket tidak terhubung: ", i)
                text.insert("insert", "({}) : {}\n".format(now, respond))
        except BrokenPipeError:
            text.insert("insert", "({}) : Client has been disconnected.\n".format(now))
 
 
s1 = Server()
# List of hexadecimal values
 
def sendHex(i):
    binary_data = bytes.fromhex(hex_data)
    i.sendall(binary_data)
    print(binary_data)
def connect():
    t1 = threading.Thread(target=s1.connect)
    t1.start()    
    
def ON():
    global a
    a = 0
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
  
def OFF():
#     text.delete("1.0", "end")
    global a
    a = 1
    t2 = threading.Thread(target=s1.send)
    t2.start()
 
def SclrPlus():
    global a
    global valueS
 
    if valueS < 100:
        valueS += 1
        w2.set(valueS)
 
    a = 2
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
def SclrMin():
    global a
    global valueS
    
    try:
        valueS -= 1
        w2.set(valueS)
    except Exception as e:
        print("Terjadi pengecualian:", e)
        
    a = 3
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
def OfstPlus():
    global a
    global valueO
    if valueO < 100:
        valueO += 1
        w1.set(valueO)
    a = 4
    t2 = threading.Thread(target=s1.send)
    t2.start()
    
def OfstMin():
    global a
    global valueO
    
    try:
        valueO -= 1
        w1.set(valueO)
    except Exception as e:
        print("Terjadi pengecualian:", e)
    a = 5
    t2 = threading.Thread(target=s1.send)
    t2.start()
 
def offside_value():
    value = w1.get()  # Mendapatkan nilai slider saat ini
    global a
    global valueO
 
    valueO = value
 
    a = 4
    t2 = threading.Thread(target=s1.send)
    t2.start()
    print("Nilai slider offside:", valueO)
 
def scaller_value():
    value = w2.get()  # Mendapatkan nilai slider saat ini
    global a
    global valueS
 
    valueS = value
 
    a = 2
    t2 = threading.Thread(target=s1.send)
    t2.start()
    print("Nilai slider:", valueS)
 
buttons = {} 
 
def dev1():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.1'
    print("Change Client: " + ip_client)
    d1.config(bg="green")
#     d2,d3,d4,d5,d6,d7 = buttonColors()
    default_color = root.cget("bg")
    for d in (d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d) 
 
def dev2():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.2'
    print("Change Client: " + ip_client)
    d2.config(bg="green")
    for d in (d1, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d) 
 
def dev3():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.3'
    print("Change Client: " + ip_client)
    d3.config(bg="green")
    for d in (d1, d2, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d) 
        
def dev4():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.4'
    print("Change Client: " + ip_client)
    d4.config(bg="green")
    for d in (d1, d2, d3, d5, d6, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d)
 
def dev5():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.5'
    print("Change Client: " + ip_client)
    d5.config(bg="green")
    for d in (d1, d2, d3, d4, d6, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d)
 
def dev6():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.6'
    print("Change Client: " + ip_client)
    d6.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d)
 
def dev7():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.7'
    print("Change Client: " + ip_client)
    d7.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d)
 
def dev8():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.8'
    print("Change Client: " + ip_client)
    d8.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d7, d9, d10, d11, d12, d13, d14):
            buttonColors(d)
 
def dev9():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.9'
    print("Change Client: " + ip_client)
    d9.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d7, d8, d10, d11, d12, d13, d14):
            buttonColors(d)
def dev10():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.10'
    print("Change Client: " + ip_client)
    d10.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d7, d8, d9, d11, d12, d13, d14):
            buttonColors(d)
 
def dev11():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.11'
    print("Change Client: " + ip_client)
    d11.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d12, d13, d14):
            buttonColors(d)
 
def dev12():
    global ip_client
    global dev
    dev = 1
    ip_client = '172.16.3.12'
    print("Change Client: " + ip_client)
    d12.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d13, d14):
            buttonColors(d)
 
def All():
    global dev
    dev = 0
    print ("ALL client")
    d13.config(bg="green")
    for d in (d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d14):
            buttonColors(d)
 
def statDev():
    global dev
    global a
    a = 6
    dev = 0
    print ("Check client")
    t2 = threading.Thread(target=s1.send)
    t2.start()
    for d in (d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13, d14):
            buttonColors(d)
            
def clear(): 
    text.delete("1.0", "end")
 
    
def destroy():
    root.destroy()
    exit()
 
 
if __name__ == "__main__":
    b1.configure(command=connect)
    b2.configure(command=ON)
    b3.configure(command=OFF)
    b4.configure(command=SclrPlus)
    b5.configure(command=SclrMin)
    b6.configure(command=OfstPlus)
    b7.configure(command=OfstMin)
    b8.configure(command=clear)
    d1.configure(command=dev1)
    d2.configure(command=dev2)
    d3.configure(command=dev3)
    d4.configure(command=dev4)
    d5.configure(command=dev5)
    d6.configure(command=dev6)
    d7.configure(command=dev7)
    d8.configure(command=dev8)
    d9.configure(command=dev9)
    d10.configure(command=dev10)
    d11.configure(command=dev11)
    d12.configure(command=dev12)
    d13.configure(command=All)
    d14.configure(command=statDev)
    offside_button.configure(command=offside_value)
    scaller_button.configure(command=scaller_value)
    t0 = threading.Thread(target=root.mainloop)
    t0.run()
```
